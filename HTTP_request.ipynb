Extracting information from HTML
using beautiful soup and requests

```python
import bs4

soup = bs4.BeautifulSoup(response.text, 'html.parser')

print(soup.select('meta[name="description"]')[0]['content'])
```

links on the platzys web page

With beautiful soup we have to know the structure of the page

```python
courses_link = soup.select('.Categories-category-item')
schools = [school['href'] for school in courses_link]

for school in schools:
    print(school)
```
